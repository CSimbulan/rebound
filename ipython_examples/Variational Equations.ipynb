# Variational Equations

**First order Variational Equations**

Variational equations are an essential tool in orbital mechanics. They can be used to measure the sensitivity of the system with respect to the initial conditions. For any differential equation of the form
$$\dot{r}(t) = F(r(t),t)$$
one can define a corresponding variational equation
$$\dot{\delta r}(t) = \frac{\partial F(r(t),t)}{\partial r} \delta r(t)$$
where $\delta r(t)$ is the variation of the trajectory $r(t)$. In the case of an N-body integration, the vector $r(t)$ contains the positions and velocities of all particles and has thus dimensions of $6N$. 

REBOUND supports variational equations when using the leapfrog, IAS15 and WHFast integrator. In the case of WHFast, the variational equations are integrator using a symplectic tangent map (Mikkola & Innanen 1999, Rein & Tamayo 2015).

This tutotial shows you how to use variational equations. Let us first setup a standard N-body simulation involving two planets.

```python
import rebound
import numpy as np
def create_sim():
    sim = rebound.Simulation()
    sim.add(m=1.)
    sim.add(m=1e-3, a=1.0)
    sim.add(m=1e-3, a=1.76)
    sim.move_to_com()
    #sim.integrator = "leapfrog"
    return sim
sim = create_sim()
```

Next, we add a set of variational particles to REBOUND. This is done with the following command

```python
var_i = sim.add_variational()
print(var_i)
```

The `add_variational()` function returns an integer which corresponds to the index in the particle array of the first variational particle. Let us now print all particle which are currently in the simulation:

```python
for p in sim.particles:
    print(p)
```

The first three partciles are the original particles, the last three are the variational particles. Note that all variational particles have been initialized to 0 (for both their mass and their position). We need to set at least one of the coordinates to a finite value to see any effect (the variational equations are linear and an initial condition of $\delta r=0$ will stay $0$ for all $t$). In our case, let's move one planet in the x direction. Note that the magnitude of this initial value does not matter (once again, because the variational equations are linear). 

```python
sim.particles[var_i+1].x = 1.
```

We can now integrate the system forward in time. Here, we integrate it for one orbit of the inner planet, i.e. until $t=2\pi$ and then print out the particle positions.

```python
sim.integrate(2.*np.pi)
for p in sim.particles:
    print(p)
```

Note that even though only one variational particle was initially perturbed, after a finite time $t$, this perturbation is visible in all variational particles.

To get a better idea of what these variational equations describe, let us integrate the same system again, but this time for a longer time and let us also plot the position of the inner planet and it's corresponding variational particle while we integrate.

```python
sim = create_sim()
var_i = sim.add_variational()
sim.particles[var_i+1].x = 1.
Noutput = 1000
times = np.linspace(0,2.*np.pi*10.,Noutput)
pos_real = np.zeros((Noutput,2))
pos_var  = np.zeros((Noutput,2))
for i,t in enumerate(times):
    sim.integrate(t)
    pos_real[i] = (sim.particles[1].x, sim.particles[1].y)
    pos_var[i] = (sim.particles[var_i+1].x, sim.particles[var_i+1].y)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
fig,axs = plt.subplots(1,2,figsize=(10,5))
axs[0].set_xlim(-1.2,1.2)
axs[0].set_ylim(-1.2,1.2)
axs[0].plot(pos_real[:,0],pos_real[:,1])
axs[1].plot(pos_var[:,0],pos_var[:,1]);
```

The left plot shows the orbit of the inner planet. Since the perturbations from the outer planet are small, it is almost on a perfect Keplerian orbit. The right plot shows the evolution of the variational particle corresponding to the inner particle. We start the variational particle at $x=1$ and can see that the particle slowly moves away from its starting point on an outward spiral. Note that the spiral is very regular, an indication of non-chaotic motion.

Let us now estimate where the inner planet would be at the end of the simulation, had it been shifted by a small amount in the $x$ direction, $\Delta x$, at the beginning of the simulation. We do this in two different ways. First, we will simply integrate the system again with the planet being shifted at the beginning:

```python
DeltaX = 0.0001
sim = create_sim()
sim.particles[1].x += DeltaX
sim.integrate(2.*np.pi)
print("x, y position using direct integration: %f, %f" % (sim.particles[1].x, sim.particles[1].y))
```

Next, let us calculate the same by using the variational equations. We use the formula:

$$r(t) \approx r(t)\big|_{\Delta x=0} + \Delta x \, \delta r(t)\big|_{\Delta x=0}$$

In this case we do not shift the original particle, but integrate the variational equations and add these at the end of the simulation:

```python
sim = create_sim()
var_i = sim.add_variational()
varscale= 0.000001
sim.particles[var_i+1].x = varscale  # Offset variational particle
sim.integrate(2.*np.pi)
print("x, y position using variational equations: %f, %f" %(sim.particles[1].x+sim.particles[var_i+1].x*DeltaX/varscale, sim.particles[1].y+sim.particles[var_i+1].y*DeltaX/varscale))
```

The final positions almost agree! The second approach using the variational equations has one important advantage. We only need to know the initial offset at the very end of the simulation ($\Delta x$ is only used in the last line). This allows us to calculate immediately where the final position of the planet would be had we started it with an offset twice as large:

```python
DeltaX *= 2.
print(sim.particles[1].x+sim.particles[var_i+1].x*DeltaX/varscale, sim.particles[1].y+sim.particles[var_i+1].y*DeltaX/varscale)
```

We can do this trick not only with an offset in one direction, but with an offset in multiple directions, by adding multiple variational equations.

```python
sim = create_sim()
varx_i = sim.add_variational()
vary_i = sim.add_variational()
varscale= 0.000001
sim.particles[varx_i+1].x = varscale  # Offset variational particle in x
sim.particles[vary_i+1].y = varscale  # Offset variational particle in y 
sim.integrate(2.*np.pi)
DeltaX = 0.001
DeltaY = 0.001
print(
    sim.particles[1].x+sim.particles[varx_i+1].x*DeltaX/varscale+sim.particles[vary_i+1].x*DeltaY/varscale, 
    sim.particles[1].y+sim.particles[varx_i+1].y*DeltaX/varscale+sim.particles[vary_i+1].x*DeltaY/varscale)
```

**Second order Variational Equations**

Let us now look at second order variational equations. We denote the second order variational equations with $\delta r^{(2)} (t)$. The second order equations couple to the first order equations and satisfy the following differential equation: 

$$\dot{\delta r^{(2)} }(t) = \frac{\partial^2 F(r(t),t)}{\partial^2 r} \left(\delta r(t)\right)^2  + \frac{\partial F(r(t),t)}{\partial r} \delta r^{(2)}(t)$$

We first add the first order variational equations as above.

```python
sim = create_sim()
var_i = sim.add_variational()
sim.particles[var_i+1].x = 1.
```

We then initialize the second order variational particles. To do that, we need to explicitally say to which first order variations the second order ones corresponds. If multiple parameters are varied, say $M$, then one has $M$ sets of first order variational equations. There will thus be $M^2$ second order variational equations (the second derivate in multiple dimensions is a matrix). 

```python
var_ii = sim.add_variational(order=2, index_1st_order_a=var_i, index_1st_order_b=var_i)
```

In our test case, there is only one variation that we consider. We thus have 1 second order system of variational equations and 9 particles in total.

```python
for p in sim.particles:
    print(p)
```

Note that we did not initialize the second order variational particles (they are all set to 0). This is because the parameter we vary (the x coordinate) is not linear. If we had varied the semi-major axis instead, we would have to initialize the second order variational particles consistantly. 

Let us now integrate the system forward in time as before and plot the particle's orbit, the first order variational particle's orbit and the second order variational particle's orbit.

```python
Noutput = 1000
times = np.linspace(0,2.*np.pi*10.,Noutput)
pos_real = np.zeros((Noutput,2))
pos_var1  = np.zeros((Noutput,2))
pos_var2  = np.zeros((Noutput,2))

for i,t in enumerate(times):
    sim.integrate(t)
    pos_real[i] = (sim.particles[1].x, sim.particles[1].y)
    pos_var1[i] = (sim.particles[var_i+1].x, sim.particles[var_i+1].y)
    pos_var2[i] = (sim.particles[var_ii+1].x, sim.particles[var_ii+1].y)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
fig,axs = plt.subplots(1,3,figsize=(15,5))
axs[0].set_xlim(-1.2,1.2)
axs[0].set_ylim(-1.2,1.2)
axs[0].plot(pos_real[:,0],pos_real[:,1])
axs[1].plot(pos_var1[:,0],pos_var1[:,1]);
axs[2].plot(pos_var2[:,0],pos_var2[:,1]);
```

We can use the second order variational equation to have a more accurate prediction of where the particle will be if we vary the initial x position by $\Delta x$. The formula we use is

$$r(t) \approx r(t)\big|_{\Delta x=0} + \Delta x \, \delta r(t)\big|_{\Delta x=0} + \frac12 \Delta x^2 \, \delta r^{(2)}(t)\big|_{\Delta x=0}$$



```python
DeltaX = 0.0001
print("First order estimate (x,y)  = (%f, %f)" %(sim.particles[1].x+sim.particles[var_i+1].x*DeltaX, sim.particles[1].y+sim.particles[var_i+1].y*DeltaX))
print("Second order estimate (x,y) = (%f, %f)" %(sim.particles[1].x+sim.particles[var_i+1].x*DeltaX+sim.particles[var_ii+1].x*DeltaX**2/2., sim.particles[1].y+sim.particles[var_i+1].y*DeltaX+sim.particles[var_ii+1].y*DeltaX**2/2.))
```

The second order estimate is much closer to the true solution:

```python
sim = create_sim()
sim.particles[1].x += DeltaX
sim.integrate(2.*np.pi*10.)
print("True solution (x,y) = (%f, %f)" %(sim.particles[1].x, sim.particles[1].y))
```

**Initializing variational particles using orbital parameters**

Simple initialization test. Note that one needs to be careful on the implications of changing one orbital parameter. I.e. changing inner planet's a will also change outer planets position unless heliocentric coordinates are used.

```python
sim = create_sim()
var_i = sim.add_variational()
var_ii = sim.add_variational(order=2, index_1st_order_a=var_i, index_1st_order_b=var_i)
sim._particles[var_i+1] = rebound.Particle(simulation=sim, primary=sim.particles[0], m=1e-3, a=1, variation="a")
sim._particles[var_ii+1] = rebound.Particle(simulation=sim, primary=sim.particles[0], m=1e-3, a=1, variation="a", variation_order=2)
```

```python
sim.integrate(100)
```

```python
DeltaX = 0.001
print("First order estimate (x,y)  = (%f, %f)" %(sim.particles[1].x+sim.particles[var_i+1].x*DeltaX, sim.particles[1].y+sim.particles[var_i+1].y*DeltaX))
print("Second order estimate (x,y) = (%f, %f)" %(sim.particles[1].x+sim.particles[var_i+1].x*DeltaX+sim.particles[var_ii+1].x*DeltaX**2/2., sim.particles[1].y+sim.particles[var_i+1].y*DeltaX+sim.particles[var_ii+1].y*DeltaX**2/2.))
sim = create_sim()
sim._particles[1] = rebound.Particle(simulation=sim, primary=sim.particles[0], m=1e-3, a=1.+DeltaX)
sim.integrate(100.)
print("True solution (x,y) = (%f, %f)" %(sim.particles[1].x, sim.particles[1].y))
```
